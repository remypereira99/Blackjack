```python
import random
```

```python
#All possible cards; used to build deck

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':1 or 11}
```

```python
class Card:
    
    #Class of every card in deck
    
    def __init__(self,suit,rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        
        return self.rank + ' of ' + self.suit
```

```python
class Deck:
    
    #Class that creates full deck of cards
    
    def __init__(self):
        
        self.all_cards = [] 
        
        #Generate full deck
        
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
                
    def shuffle(self):
        
        #Shuffles deck
        
        random.shuffle(self.all_cards)
        
    def show_cards(self):
        
        #Prints every card in order
        
        for card in self.all_cards:
            print(card)
            
    def deal(self):
        
        #Removes top card from deck
        
        return self.all_cards.pop(0)
```

```python
class Player:
    
    #Class for human player
    
    def __init__(self):
        
        #Attribute for player's cards
        
        self.cards = []
        
        #Attribute for player's money
        
        self.bank = 250
        
    def add_cards(self,new_cards):
        
        #Adds cards to player's hand
        
        if type(new_cards) == type([]):
            self.cards.extend(new_cards)
        
        else:
            self.cards.append(new_cards)
            
    def show_hand(self):
        
        #Shows cards in player's hand and their value
        
        print(f'\nYou have {self.cards[0]} and {self.cards[1]}')
        
        global hand_value
        
        hand_value = 0
        
        for card in self.cards:
            hand_value += card.value
        
        print(f'\nTheir value is {hand_value}')
            
    def wager(self):
        
        #Places wager
        
        global placed_wager 
        
        placed_wager = input('\nHow much would you like to wager? (Minimum £5)')
        
        try:
            if int(placed_wager) > self.bank:
                print('\nYou do not have enough money! Please try again.')
                self.wager()
            elif int(placed_wager) > 5:
                self.bank -= int(placed_wager)
                print(f'\nYou have placed a wager of £{placed_wager}. You have £{self.bank} remaining')
            elif int(placed_wager) < 5:
                print('\nWager too low! Please try again.')
                self.wager()
        
        #Error handling in case input is not integer
        
        except:
            print('\nSomething went wrong! Please try again.')
            self.wager()
        
```
