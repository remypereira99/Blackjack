```python
#Variable for determining whether game has been played before

global game_on

game_on = False

values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8,
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':1}
```

```python
class Card:
    
    #Class of every card in deck
    
    def __init__(self,suit,rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        
        return self.rank + ' of ' + self.suit
    
        #e.g. 'Two of Hearts'
```

```python
class Deck:
    
    #Class that creates full deck of cards
    
    def __init__(self):
        
        suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
        ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
        values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8,
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':1}
        
        self.all_cards = [] 
        
        #Generate full deck
        
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
                
    def shuffle(self):
        
        #Shuffles deck
        
        import random
        
        random.shuffle(self.all_cards)
        
    def show_cards(self):
        
        #Prints every card in order
        
        for card in self.all_cards:
            print(card)
            
```

```python
class Player:
    
    def __init__(self):
        
        #Attributes for cards in hand, money in bank, placed wager, and value of cards in hand
        
        self.cards = []
        
        self.bank = 250
        
        self.placed_wager = 0
        
        self.hand_value = 0
        
    def add_cards(self,new_cards):
        
        #Adds card(s) to player's hand
        
        if type(new_cards) == type([]):
            self.cards.extend(new_cards)
            for card in new_cards:
                self.hand_value += card.value
        
        else:
            self.cards.append(new_cards)
            self.hand_value += new_cards.value
            
    def show_hand(self):
        
        #Shows cards in player's hand and their value
            
        for num in range(len(self.cards)):
            
            print(f'You have {self.cards[num]}.')
        
        print(f'\nTheir total value is {self.hand_value}.')
            
    def wager(self):
        
        #Asks how much money you want to wager
        
        self.placed_wager = int(input(f'\nYou have {self.bank}. How much would you like to wager? (Minimum £5)'))
        
        try:
            if self.placed_wager > self.bank:
                print('\nYou do not have enough money! Please try again.')
                self.wager()
            elif self.placed_wager >= 5:
                self.bank -= self.placed_wager
                print(f'\nYou have placed a wager of £{self.placed_wager}. You have £{self.bank} remaining')
            elif self.placed_wager < 5:
                print('\nWager too low! Please try again.')
                self.wager()
        
        #Error handling in case input is not integer
        
        except:
            print('\nSomething went wrong! Please try again.')
            self.wager()
            
    def clear_hand(self):
        
        #Resets hand
        
        self.cards = []
        
        self.hand_value = 0
```

```python
class Dealer:
    
    #Class for CPU dealer
    
    def __init__(self):
        
        #Attribute for player's cards
        
        self.cards = []
        
        self.hand_value = 0
        
    def add_cards(self,new_cards):
        
        #Adds cards to player's hand
        
        if type(new_cards) == type([]):
            self.cards.extend(new_cards)
            for card in new_cards:
                self.hand_value += card.value
        
        else:
            self.cards.append(new_cards)
            self.hand_value += new_cards.value
            
    def show_hand(self):
        
        #Shows cards in player's hand and their value
            
        for num in range(len(self.cards)):
            
            print(f'Dealer has {self.cards[num]}.')
        
        print(f'\nTheir total value is {self.hand_value}.')
```

```python
def card_value_check(player):
        
        #Function that checks whether player has blackjack or has gone bust
        
        if player.hand_value == 21:
        
            print(f'\nBlackjack! You win {4 * player.placed_wager}.')
        
            player.bank += 2 * player.placed_wager
        
            player.placed_wager = 0
            
            replay_message()
            
        elif player.hand_value > 21:
    
            print(f'\nBust! You went over 21. You lose £{player.placed_wager}. You have £{player.bank} remaining')
        
            player.placed_wager = 0
            
            replay_message()
            
        else:
            if hit > 0:
                hit_stay(player1)
```

```python
def game_over(player,dealer):
    
    #Function that ends the game and determines the winner
    
    if dealer.hand_value > player.hand_value:
        
        if dealer.hand_value < 22:
        
            print('Dealer Wins.')
        
            print(f'You lose {player.placed_wager}. You have £{player.bank} remaining.')
        
            replay_message()
        
        else:
            
            print('Dealer went bust. You win!')
            
            print(f'You receive {2 * player.placed_wager}!')
        
            player.bank += 2 * player.placed_wager
        
            player.placed_wager = 0
    
            replay_message()
                      
    elif dealer.hand_value < player.hand_value:
                      
        print('You win!')
        
        print(f'You receive {2 * player.placed_wager}!')
        
        player.bank += 2 * player.placed_wager
        
        player.placed_wager = 0
    
        replay_message()
```

```python
def replay_message():
    
    #Function that allows you to choose if you want to replay or not
    
    replay = input('WOULD YOU LIKE TO PLAY AGAIN? (Yes/No)')

    if replay.upper() == 'YES':
        
        game_logic()
        
    elif replay.upper() == 'NO':
        
        print('\nTHANKS FOR PLAYING!')
        
        global game_on
        
        game_on = False
```
